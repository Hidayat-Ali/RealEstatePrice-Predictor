```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
# Load the dataset here 
df1 = pd.read_csv("Data/Bengaluru_House_Data.csv")
df1.head()
```

```python

df1.shape
df1.groupby('area_type')['area_type'].agg('count')
```

```python
# drop some features that are not required in the model
df2 = df1.drop(['area_type','society','balcony','availability'],axis="columns")
df2.head()
```

## NOW LETS START THE DATA CLEANING PROCESS


```python
df2.isnull().sum()
# since my dataset is bit huge and the null values are very less as compared to the dataset 
# so it is better to drop them or we can use other approach to fill the missing values like mean,medain etc 
```

```python
df3 = df2.dropna()
df3.isnull().sum()
df3.head()
```

```python
df3["size"].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x :int(x.split(' ')[0]))
df3.head()
```

```python
def isFloat(x):
    try:
        float(x)   
        return True
    except:
        return False

```

```python
df4 = df3[~df3['total_sqft'].apply(isFloat)].head(10)
df4
```

```python
def convert_sqft_to_num(x):
   tokens = x.split('-')
   if len(tokens) ==2:
      return (float(tokens[0])+float(tokens[1]))/2
   try:
      return float(x)
   except:
      None
```

```python
df4['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
df4
```
